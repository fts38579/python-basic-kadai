```python
# 条件分岐のif文の課題

#ランダムな整数を利用するために、randomモジュールをインポート
import random

# 変数varに0～15までのランダムな整数を代入する
var = random.randint(0,15)

# 変数varの値を出力
print(var)

# ３の倍数かつ５の倍数であるとき
if  var % 15== 0:
    print("FizzBuzz")

# ５の倍数であるとき
elif var % 5 == 0:
    print("Buzz")

# ３の倍数であるとき
elif var % 3 == 0:
    print("Fizz")

# 該当しないとき
else:
    print(var)
```
